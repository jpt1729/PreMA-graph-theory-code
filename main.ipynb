```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
class Graph:
    def __init__(self, vertices = [], edges = []):
        self.vertices = vertices
        self.edges = edges
        self.x = [vertex[0] for vertex in vertices]
        self.y = [vertex[1] for vertex in vertices]
        
    def plot(self, padding = 1):
        fig, ax = plt.subplots()
        for vertex in self.vertices:
            ax.plot(vertex[0], vertex[1], 'o')
        for edge in self.edges:
            ax.plot([self.vertices[edge[0]][0], self.vertices[edge[1]][0]],
                    [self.vertices[edge[0]][1], self.vertices[edge[1]][1]], '-')
        ax.set_xlim([min(self.x) - padding, max(self.x) + padding])
        ax.set_ylim([min(self.y) - padding, max(self.y) + padding])
        plt.show()
        return 0

    def save_plot(self, padding = 1, **kwargs):
        '''save_plot(self, padding = 1, **kwargs)
        Every other option is used as input into plt.savefig
        Saves figure as file
        '''
        fig, ax = plt.subplots()
        for vertex in self.vertices:
            ax.plot(vertex[0], vertex[1], 'o')
        for edge in self.edges:
            ax.plot([self.vertices[edge[0]][0], self.vertices[edge[1]][0]],
                    [self.vertices[edge[0]][1], self.vertices[edge[1]][1]], '-')
        ax.set_xlim([min(self.x) - padding, max(self.x) + padding])
        ax.set_ylim([min(self.y) - padding, max(self.y) + padding])
        
        plt.savefig(**kwargs)
        return 0
```

```python
# Define the vertices and edges of the graph
vertices = [(2, 2), (1, 1), (2, 1), (1, 2)]
edges = [(0, 1), (1, 2), (2, 3), (3, 0)]

# Create a new figure and axis
fig, ax = plt.subplots()

# Plot the vertices and edges
for vertex in vertices:
    ax.plot(vertex[0], vertex[1], 'o')
for edge in edges:
    ax.plot([vertices[edge[0]][0], vertices[edge[1]][0]],
            [vertices[edge[0]][1], vertices[edge[1]][1]], '-')

# Set the axis limits and show the plot
ax.set_xlim([-1, 4])
ax.set_ylim([-1, 4])
plt.show()
```

```python
def create_graph(vertices=[], edges=[], starting_point=1, num_vertices=15):
    # creates a ton of points + edges
    new_vertices = []
    new_edges = []
    x, y = vertices[starting_point]
    for i in range(num_vertices):
        # Graph Points
        new_vertices.append((x + num_vertices/2 - abs((y + num_vertices/2 - i)*0.2) + random.randint(5,10) * 0.1, y + num_vertices/2 - i))
        new_edges.append((starting_point, len(
            vertices) + len(new_vertices) - 1))

    return (vertices + new_vertices, edges + new_edges)

example_graph_vertices = [(0, 0), (5, 0)]
example_graph_edges = [(0, 1), (7, 4)]
example_graph_vertices, example_graph_edges = create_graph(
    vertices=example_graph_vertices, edges=example_graph_edges, num_vertices = 15)

example_graph = Graph(vertices=example_graph_vertices,
                      edges=example_graph_edges)

example_graph.plot(padding=1)
example_graph.save_plot(padding=1, fname = 'figures/example.png')
```
